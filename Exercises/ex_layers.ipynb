```python
import numpy as np
from si.neural_networks.layers import DenseLayer, Dropout
from si.neural_networks.optimizers import SGD
```

```python
np.random.seed(42)

X = np.random.rand(5, 3)   # 5 amostras, 3 features
y = np.random.rand(5, 2)   # target fict√≠cio
```

```python
optimizer = SGD(learning_rate=0.01)
```

```python
dense = DenseLayer(n_units=4, input_shape=(3,))
dense.initialize(optimizer)

dropout = Dropout(probability=0.5)
```

```python
print("=== FORWARD (training=True) ===")
out_dense = dense.forward_propagation(X, training=True)
out_dropout = dropout.forward_propagation(out_dense, training=True)

print("Output Dense:\n", out_dense)
print("Output Dropout:\n", out_dropout)
```

```python
print("\n=== BACKWARD ===")
fake_error = np.random.rand(*out_dropout.shape)

error_dropout = dropout.backward_propagation(fake_error)
error_dense = dense.backward_propagation(error_dropout)

print("Error after Dropout:\n", error_dropout)
print("Error after Dense:\n", error_dense)
```

```python
print("\n=== FORWARD (training=False) ===")
out_dense_test = dense.forward_propagation(X, training=False)
out_dropout_test = dropout.forward_propagation(out_dense_test, training=False)

print("Output Dense:\n", out_dense_test)
print("Output Dropout (no dropout):\n", out_dropout_test)
```
